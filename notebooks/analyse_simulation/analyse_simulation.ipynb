```python
# import the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn


# Load the data
df = pd.read_csv("C:/Users/39351/warehouse-simulation-logistics/data/processed/order_log.csv")

df.head()
```

## Summary Statistics

```python
df[["waiting_time", "service_time"]].describe()
```

## Plot Distribution of Wait Time

```python
plt.figure(figsize=(8, 4))
sn.histplot(df['waiting_time'], bins=20, kde=True, color="orange")
plt.title("Distribution of waiting time")
plt.ylabel("Frequency")
plt.xlabel("Wait Time (Minutes)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Plot Service Time Distribution

```python
plt.figure(figsize=(8, 4))
sn.histplot(df['service_time'], bins=20, kde=True, color='Blue')
plt.title('Service Time Distribution')
plt.ylabel('Frequency')
plt.xlabel('Service Time (Minutes)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Wait Time over Time (Trend)

```python
plt.figure(figsize=(8, 4))
sn.lineplot(data=df, x="arrival_time", y="waiting_time", color='brown')
plt.title("Wait Time Over Time")
plt.xlabel("Arrival Time (Minutes)")
plt.ylabel("Wait Time")
plt.grid(True)
plt.show()
```

# Load the data of different scenarios

```python
df_all = pd.read_csv("C:/Users/39351/warehouse-simulation-logistics/data/processed/all_scenarios.csv")

#df_all.head()
```

# Average Wait and  Service Time by Scenarios

```python
df_all.groupby('scenario_label')[['waiting_time', 'service_time']].mean(). round(2)
```

# Compare Wait Times by Scenario

```python
plt.figure(figsize=(8, 4))
sn.boxenplot(data=df_all, x = 'scenario_label', y= 'waiting_time', color='purple')
plt.title("Wait Time comparison by Scenarios")
plt.ylabel("Wait Time (Minutes)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
